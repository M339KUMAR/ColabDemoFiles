##8. NEWTON 2nd Law

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

g=9.81      #m/s^2

def N2M(x, t, M, c):
       x[0], x[1] = x
       F = 10  # Newtons
       dxdt = [x[1], ((F*x[0])/(M*g))-c]
       return   dxdt

Xint = [1, 1]    #initial condition of the system
M = 10        #Inertial mass of the body
c =0      # assuming smooth frictionless surface
t = np.linspace(0, 5, 101) # time interval

Vel, Acc= odeint(N2M, Xint, t, args=(M,c), Dfun= None, col_deriv=0, full_output=0, ml=None, mu= None, rtol= None, atol= None, tcrit= None, tfirst=False).T
print(Vel, Acc)

plt.plot(t, Vel, 'Blue', label= 'F= ma', marker= 'o', markersize='4', linestyle='--')
plt.plot(t, Acc, 'Red', label= 'dx/dt=(F/(M*g))*x', marker='x', markersize= '6', linestyle='-')
plt.legend(loc='best')
plt.xlabel('Time')
plt.ylabel(' "Vel (m/sec) & Acc( m/sec^2 )" of the body:  dx/dt')
plt.tight_layout()
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0)),
plt.grid()
plt.show()
```

##9.  SHM of Pendulum

```python
import numpy as np

def pend(y, t, b, c):
       theta, omega = y
       dydt = [ omega, -b*omega-c*np.sin(theta)]
       return dydt

b = 0.25
c = 5.00

y0 = [np.pi-0.1, 0.0]

t = np.linspace(0, 50, 101)

from scipy.integrate import odeint

Avel, Aacc = odeint(pend, y0, t, args= (b, c)).T
print( Avel)
print("***************************************")
print("Num of calculations  :", len(Avel), len(Aacc))
print("***************************************")
print (Aacc )
import matplotlib.pyplot as plt
plt.plot(t, Avel, 'Red', label = 'Ang Displacement')
plt.plot(t, Aacc, 'Orange', label = 'Ang Velocity')
#plt.plot(t, omega, 'r', label = 'Ang Acc')
plt.legend(loc= 'best')
plt.xlabel("Time")
plt.ylabel(" rad/sec, rad/s^2")
plt.grid()
plt.savefig("SHM.png")
plt.show()
```

```python
data = [
    {"name": "alice smith", "age": 30, "email": "alice@example.com", "salary": 50000.00, "join_date": "2022-03-15"},
    {"name": "bob gray", "age": 17, "email": "bob@not-an-email", "salary": 60000.00, "join_date": "invalid-date"},
    {"name": "charlie brown", "age": None, "email": "charlie@example.com", "salary": -1500.00, "join_date": "15-09-2022"},
    {"name": "dave davis", "age": 45, "email": "dave@example.com", "salary": 70000.00, "join_date": "2021-07-01"},
    {"name": "eve green", "age": 25, "email": "eve@example.com", "salary": None, "join_date": "2023-12-31"},
]
```

```python
#1 Capitalize Strings

# Capitalizing the names for consistency
data = [{**d, "name": d["name"].title()} for d in data]
print(type(data[0]['age']))
```

```python
#2 Convert Data Types

# Converting age to an integer type, defaulting to 25 if conversion fails
data = [{**d,"age": int(d["age"]) if isinstance(d["age"], (int, float)) else 25} for d in data]
print(type(data))
print(type(data[0]['age']))
```

```python
#3 Validate Numeric Ranges

# Ensuring age is an integer within the range of 18 to 60; otherwise, set to 25
data = [{**d, "age": d["age"] if isinstance(d["age"], int) and 18 <= d["age"] <= 60 else 25} for d in data]
print(type(data[0]['age']))
```

```python
#4 Verifying that the email contains both an "@" and a ".";

#assigning 'invalid@example.com' if the format is incorrect
data = [{**d, "email":d["email"] if "@" in d["email"] and "." in d["email"] else "invalid@example.com"} for d in data]
print(type(data[0]['email']))
```

```python
#5 Handle Missing Values

# Assigning a default salary of 30,000 if the salary is missing
data = [{**d, "salary":d["salary"] if d["salary"] is not None else 30000.00} for d in data]
print(type(data[0]['join_date']))
```

```python
from datetime import datetime

#6 Attempting to convert the date
# to a standardized format and
# defaulting to '2023-01-01'
# if invalid
data = [{**d, "join_date": (lambda x: (datetime.strptime(x, '%Y-%m-%d').date() if '-' in x and len(x) == 10 else datetime.strptime(x, '%d-%m-%Y').date() if x and 'invalid-date' not in x else '2023-01-01')(d['join_date']))} for d in data]
print(type(data[0]['join_date']))
print(data[0]['join_date'])
```

```python
#7 Remove Negative Values

# Replacing negative salary values with zero to ensure all values are non-negative
data = [{**d, "salary": max(d["salary"], 0)} for d in data]
data[2]#['salary']
print(type(data[0]['join_date']))
```

```python
#8 Check for Duplicates

# Keeping only unique entries based on the name field
data = {tuple(d.items()) for d in data}  # Using a set to remove duplicates
data = [dict(t) for t in data]  # Converting back to list of dictionaries
#display(data1)
#print('*'*30)
#display(data2)
print(type(data[0]['join_date']))
```

```python
#9 Scale Numeric Values

# Normalizing salary values to a percentage of the maximum salary
max_salary = max(d["salary"] for d in data)
data = [{**d, "salary": (d["salary"] / max_salary * 100) if max_salary > 0 else 0} for d in data]
data
print(type(data[0]['join_date']))
```

```python
#10 Trim Whitespaces

# Trimming whitespace from names for cleaner data
data = [{**d, "name": d["name"].strip()} for d in data]
for i in range(len(data)):
   print(data[i]['name'])
```

# METHOD RESOLUTION ORDER

```python
class A:
   a = 'Anitha'
   print('Class A:', a)

class B(A):  #B inherits Class A
   b = 'Bavana'
   print('Class B:', b)

class C(B):  #C inherits Class A
   c = 'Chinni'
   print('Class C:', c)

class D(C, B):
# Here order of inheritance is important

# C, B works
#Just inheriting Class C also works
#because C inherits Class B
#-------------------
#but not B, C
#TypeError: Cannot create a
#consistent method resolution
#order (MRO) for bases B, C
#-------------------
   d = 'David'
   print('Class D:', d)

"""
     METHOD RESOLUTION ORDER
1.each ancestor class appears exactly
once

2.a class always appears before
its ancestor ("monotonicity")

3.direct parents of the same class
should appear in the same order
as they are listed in class definition
("consistent local precedence order")

4.if children of class A always appear
before children of class B, then A
should appear before B
("consistent extended
precedence order")

"""

Duo = D()
print('\n')
print(dir(Duo))
print(Duo)
print('\n')
print(Duo.__class__)
print(Duo.__dict__)
print(Duo.__dir__)
print(Duo.__doc__)
print(Duo.__format__)
print(Duo.__getattribute__)
print(Duo.__hash__)
print(Duo.__init__)
print(Duo.__module__)
print(Duo.__ne__)
print(Duo.__new__)
print(Duo.__reduce__)
print(Duo.__repr__)
print(Duo.__sizeof__)
print(Duo.__init_subclass__)
print()
print(Duo.a,'\n',Duo.b,'\n',Duo.c,'\n',Duo.d)

print()
print('hash:',hash(Duo))
print()

print(format(Duo))

print(hash(Duo))

print(repr(Duo))
```

```python
class A:
   a = 'Anitha'
   print('Class A:', a)

class B(A):   # B inherits Class A
   b = 'Bavana'
   print('Class B:', b)

class C(A):   # C inherits Class A
   c = 'Chinni'
   print('Class C:', c)

class D(B, C):
# B, C and C, B both work just fine
   d = 'David'
   print('Class D:', d)

Dobj = D()
print(Dobj)
print(Dobj.a)
print(Dobj.b)
print(Dobj.c)
print(Dobj.d)
```

```python
import logging
logging.basicConfig(filename='example.log', level=logging.INFO)


def logger(func):
    def log_func(*args):
        logging.info(
            'Running "{}" with arguments {}'.format(func.__name__, args))
        print(func(*args))
    return log_func

def add(x, y):
    return x+y


def sub(x, y):
    return x-y

add_logger = logger(add)
sub_logger = logger(sub)

add_logger(3, 3)
add_logger(4, 5)

sub_logger(10, 5)
sub_logger(20, 10)
```
