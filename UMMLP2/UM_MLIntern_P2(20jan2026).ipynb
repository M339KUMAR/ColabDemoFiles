```python

import os

def count_files_os(folder_path):
    """Counts the number of files in a given folder using os.listdir."""

    count = 0

    # List all entries in the directory
    for file in os.listdir(folder_path):

        # Construct the full path
        full_path = os.path.join(folder_path, file)

        # Check if the entry is a file (and not a sub-directory or symlink)
        if os.path.isfile(full_path):
            count += 1

    return count

# Example Usage:
# Replace 'your/folder/path' with the actual path to your folder
folder_to_check = '/content/FD_P2/data'

file_count = count_files_os(folder_to_check)
#print(f"Total files count ::'{folder_to_check}': {file_count}")
print(f"Total files count ::: {file_count}")
```

###Combine all 183 pkl files

```python

#import pandas as pd
#import glob
#import os

# Define the directory where your pickle files are stored
# Replace 'path/to/your/pickles' with the actual directory path
pickles_dir = '/content/FD_P2/data'    #'path/to/your/pickles'

# Use glob to find all files ending with .pkl in the specified directory
# This creates a list of file paths to iterate through
pickle_files = glob.glob(os.path.join(pickles_dir, '*.pkl'))

# Check if any files were found
if not pickle_files:
    print(f"No pickle files found in the directory: {pickles_dir}")
else:
    # Initialize an empty list to store individual dataframes
    list_of_dfs = []

    # Iterate over each file path in the list
    for file in pickle_files:
        try:
            # Read the pickle file into a temporary dataframe and append it to the list
            df_temp = pd.read_pickle(file)
            list_of_dfs.append(df_temp)
            print('--------------------------------------------------------')
            print(f"Successfully loaded {file} with {len(df_temp)} records.")
            print('--------------------------------------------------------')
        except Exception as e:
            # Handle potential errors during file reading
            print(f"Error reading file {file}: {e}")

    # Concatenate all dataframes in the list into a single final dataframe
    if list_of_dfs:
        final_dataframe = pd.concat(list_of_dfs, ignore_index=True)

        print("\n--- Summary ---")
        print(f"Total number of files processed: {len(pickle_files)}")
        print(f"Total records in the combined dataframe: {len(final_dataframe)}")
        print("\nCombined dataframe head:")
        print(final_dataframe.head())

        # Optional: Save the combined dataframe to a new pickle file
        # final_dataframe.to_pickle("combined_records.pkl")
        # print("\nCombined dataframe saved to 'combined_records.pkl'")

    else:
        print("No dataframes were successfully loaded. Cannot create a combined dataframe.")
```
