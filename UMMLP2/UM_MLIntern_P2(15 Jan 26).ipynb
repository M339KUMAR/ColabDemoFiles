```python

import os

def count_files_os(folder_path):
    """Counts the number of files in a given folder using os.listdir."""

    count = 0

    # List all entries in the directory
    for file in os.listdir(folder_path):

        # Construct the full path
        full_path = os.path.join(folder_path, file)

        # Check if the entry is a file (and not a sub-directory or symlink)
        if os.path.isfile(full_path):
            count += 1

    return count

# Example Usage:
# Replace 'your/folder/path' with the actual path to your folder
folder_to_check = '/content/FD_P2/data'

file_count = count_files_os(folder_to_check)
#print(f"Total files count ::'{folder_to_check}': {file_count}")
print(f"Total files count ::: {file_count}")
```

```python

import glob
import os

# Define the directory path
directory_path = '.'  # Current directory

# Pattern to match all files and directories
# '*' matches everything directly in the specified path
pattern = os.path.join(directory_path, '*')

# Use glob.glob() to get the list of files
file_list = glob.glob(pattern)

print("Files in directory:")
for file in file_list:
    print(file)
```
