```python

import os

def count_files_os(folder_path):
    """Counts the number of files in a given folder using os.listdir."""

    count = 0

    # List all entries in the directory
    for file in os.listdir(folder_path):

        # Construct the full path
        full_path = os.path.join(folder_path, file)

        # Check if the entry is a file (and not a sub-directory or symlink)
        if os.path.isfile(full_path):
            count += 1

    return count

# Example Usage:
# Replace 'your/folder/path' with the actual path to your folder
folder_to_check = '/content/FD_P2/data'

file_count = count_files_os(folder_to_check)
#print(f"Total files count ::'{folder_to_check}': {file_count}")
print(f"Total files count ::: {file_count}")
```

###Combine all 183 pkl files

```python

#import pandas as pd
#import glob
#import os

# Define the directory where your pickle files are stored
# Replace 'path/to/your/pickles' with the actual directory path
pickles_dir = '/content/FD_P2/data'    #'path/to/your/pickles'

# Use glob to find all files ending with .pkl in the specified directory
# This creates a list of file paths to iterate through
pickle_files = glob.glob(os.path.join(pickles_dir, '*.pkl'))

# Check if any files were found
if not pickle_files:
    print(f"No pickle files found in the directory: {pickles_dir}")
else:
    # Initialize an empty list to store individual dataframes
    list_of_dfs = []

    # Iterate over each file path in the list
    for file in pickle_files:
        try:
            # Read the pickle file into a temporary dataframe and append it to the list
            df_temp = pd.read_pickle(file)
            list_of_dfs.append(df_temp)
            print('--------------------------------------------------------')
            print(f"Successfully loaded {file} with {len(df_temp)} records.")
            print('--------------------------------------------------------')
        except Exception as e:
            # Handle potential errors during file reading
            print(f"Error reading file {file}: {e}")

    # Concatenate all dataframes in the list into a single final dataframe
    if list_of_dfs:
        final_dataframe = pd.concat(list_of_dfs, ignore_index=True)

        print("\n--- Summary ---")
        print(f"Total number of files processed: {len(pickle_files)}")
        print(f"Total records in the combined dataframe: {len(final_dataframe)}")
        print("\nCombined dataframe head:")
        print(final_dataframe.head())

        # Optional: Save the combined dataframe to a new pickle file
        # final_dataframe.to_pickle("combined_records.pkl")
        # print("\nCombined dataframe saved to 'combined_records.pkl'")

    else:
        print("No dataframes were successfully loaded. Cannot create a combined dataframe.")
```

#SCENARIO - 2

#### Every Day, A list of two terminals is drawn at random. All Transactions on these terminals in the next 28 days will be marked as fraudulent. This scenario simulates a criminal use of a terminal, through phishing for example. You could add features that keep track of the number of fraudulent transactions on the terminal to help with this scenario.

```python

import pandas as pd
import numpy as np

final_df2 = final_dataframe.copy(deep = True)

# Ensure datetime
final_df2['TX_DATETIME'] = pd.to_datetime(final_df2['TX_DATETIME'])
final_df2 = final_df2.sort_values('TX_DATETIME').reset_index(drop=True)

final_df2['TX_FRAUD'] = 0
final_df2['TX_FRAUD_SCENARIO'] = 0

days = final_df2['TX_TIME_DAYS'].unique()

np.random.seed(42)

for day in days:
    terminals_today = np.random.choice(
        final_df2['TERMINAL_ID'].unique(), size=2, replace=False
    )

    mask = (
        (final_df2['TX_TIME_DAYS'] >= day) &
        (final_df2['TX_TIME_DAYS'] <= day + 28) &
        (final_df2['TERMINAL_ID'].isin(terminals_today))
    )

    final_df2.loc[mask, 'TX_FRAUD'] = 1
    final_df2.loc[mask, 'TX_FRAUD_SCENARIO'] = 2
```

```python
final_df2
```

```python
final_df2[['TX_FRAUD','TX_FRAUD_SCENARIO']].value_counts()
```

```python
final_df2['TX_DATETIME'] = pd.to_datetime(final_df2['TX_DATETIME'])
```

```python

final_df2 = final_df2.sort_values(['TERMINAL_ID', 'TX_DATETIME'])

#final_df2['TX_DATETIME'] = pd.to_datetime(final_df2['TX_DATETIME'])

final_df2 = final_df2.sort_values(['TERMINAL_ID','TX_DATETIME'])

final_df2 = final_df2.set_index('TX_DATETIME')


final_df2['TERM_FRAUD_COUNT_1D'] = (
    final_df2.groupby('TERMINAL_ID')['TX_FRAUD']
    .rolling('1D')#, on=final_df2['TX_DATETIME'])
    .sum()
    .reset_index(level=0, drop=True)
)

final_df2['TERM_FRAUD_COUNT_7D'] = (
    final_df2.groupby('TERMINAL_ID')['TX_FRAUD']
    .rolling('7D')
    .sum()
    .reset_index(level=0, drop=True)
)

final_df2['TERM_TX_COUNT_7D'] = (
    final_df2.groupby('TERMINAL_ID')['TX_AMOUNT']
    .rolling('7D')
    .count()
    .reset_index(level=0, drop=True)
)

final_df2['TERM_FRAUD_RATIO_7D'] = (
    final_df2['TERM_FRAUD_COUNT_7D'] / final_df2['TERM_TX_COUNT_7D']
).fillna(0)
```

```python
final_df2.index
```

```python

features = [
    'TX_TIME_SECONDS',
    'TX_TIME_DAYS',
    'TERM_FRAUD_COUNT_1D',
    'TERM_FRAUD_COUNT_7D',
    'TERM_TX_COUNT_7D',
    'TERM_FRAUD_RATIO_7D'
]

X = final_df2[features]
y = final_df2['TX_FRAUD']
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42, stratify=y
)

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    class_weight='balanced',
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

# SCENARIO - 3

####Every Day a list of 3 customers is drawn at random. In the next 14 Days, 1/3 of their transactions have their amounts multiplied by 5 and marked as fraudulent. This scenario simulates a card-not-present fraud where the Credentials of a customer have been leaked. The customer continues to make transactions, and transactions of higher values are made by the fraudsters whivtries to maximize their gains. You could add features that keep track of the spending habits of the customer for this scenario

```python
import pandas as pd
import numpy as np

np.random.seed(42)   # for reproducibility

# ---- Example DataFrame (replace with your real df) ----
df = pd.DataFrame({
    "cust_id": np.random.randint(1, 100, 1000),
    "amount": np.random.randint(100, 1000, 1000),
    "f2": np.random.randn(1000),
    "f3": np.random.randn(1000),
    "f4": np.random.randn(1000),
    "f5": np.random.randn(1000),
    "f6": np.random.randn(1000),
})

# Add date column (if you already have one, remove this)
df["txn_date"] = pd.Timestamp.today().normalize()

# Add fraud flag column
df["fraud_flag"] = 0

# ------------------------------------------------------
# STEP 1: Select 3 random customers
# ------------------------------------------------------
selected_customers = np.random.choice(df["cust_id"].unique(), size=3, replace=False)
print("Selected Customers:", selected_customers)

df_sub = df[df["cust_id"].isin(selected_customers)]

# ------------------------------------------------------
# STEP 2: Select 1/3rd of their records
# ------------------------------------------------------
n = int(len(df_sub) / 3)
fraud_idx = np.random.choice(df_sub.index, size=n, replace=False)

# ------------------------------------------------------
# STEP 3: Assign random dates in next 14 days
# ------------------------------------------------------
start_date = pd.Timestamp.today().normalize()
random_days = np.random.randint(1, 15, size=n)
df.loc[fraud_idx, "txn_date"] = start_date + pd.to_timedelta(random_days, unit="D")

# ------------------------------------------------------
# STEP 4: Multiply amount by 5
# ------------------------------------------------------
df.loc[fraud_idx, "amount"] *= 5

# ------------------------------------------------------
# STEP 5: Mark as Fraud
# ------------------------------------------------------
df.loc[fraud_idx, "fraud_flag"] = 1

# ------------------------------------------------------
# RESULTS
# ------------------------------------------------------
print("\nTotal Fraud Records:", df["fraud_flag"].sum())
print("\nSample Fraud Records:")
print(df[df["fraud_flag"] == 1].head())
```
