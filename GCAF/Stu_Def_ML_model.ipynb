```python
#importing Neccesary Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Loading the dataset

Df = pd.read_csv('https://raw.githubusercontent.com/atulpatelDS/Data_Files/master/Credit_Card_Defaulter/Credit_Card_defaulters.csv', delimiter=',')

Df.to_csv('/content/sample_data/Stud_CC_Default.csv')
Df.to_excel('/content/sample_data/Stud_CC_Default.xlsx')
Df.shape
```

```python
#Descriptive Statistics Of Non Numeric Columns

print(Df['default'].describe(),'\n\n',Df['student'].describe())
```

```python
#Overall Descriptive Statistics of the DataFrame

Df.describe()
```

```python
#plot of default column based on male ad female category

sns.catplot(x = "default",data= Df, kind="count", height=4, aspect=1,edgecolor='k')
```

```python
# Value counts of default and student column based on category

Df_def=Df['default'].value_counts()
Df_Stu=Df['student'].value_counts()
print(Df_def,' \n',Df_Stu)
```

```python
#Drawing Prob Density Function plot of Balance

sns.kdeplot(x='balance',data=Df)
```

```python
#Drawing Emperical Cummulative Density Function of balance

sns.ecdfplot(x='balance',data=Df)
```

```python
#Drawing Prob Density Function plot of income

sns.kdeplot(x='income', data=Df)
```

```python
#Drawing Emperical Cummulative Density Function of income

sns.ecdfplot(x='income', data =Df)
```

```python
#Checking the column names of the DataFrame

print('\n', Df.columns)
```

```python

#plotting histogram of balance column in DataFrame

Df['balance'].hist(bins=50, rwidth=1.0, color='magenta')
plt.show()
```

```python
#plotting histogram of income column in DataFrame

plt.figure(figsize=(15,8))
Df['income'].hist(bins=50, rwidth=1.0, color='green')
plt.show()
```

```python

#plotting default values as bar kind on cat

Df_def.plot(kind='bar', color=['red','green'], edgecolor=['green','red'], lw=2.0)
plt.legend()
plt.show()
```

```python

#plotting student values as bar kind on cat

Df_Stu.plot(kind='bar', color=['cyan','magenta'], edgecolor=['green','red'], lw=2.0)
plt.legend()
plt.show()
```

```python

#countplot of seaborn of DataFrame

sns.countplot(Df[['default','student','balance','income']])
```

```python

#Complete Information about DataFrame

Df.info()
```

```python
#Calculating Correlation b/n Numerical Column Of DataFrame

Corr=Df[['balance','income']].corr()
Corr
```

```python
#plotting heatmap to see which are highly correlated columns

sns.heatmap(Corr,
                          vmin=-1.0, vmax=1.0,
                          cmap= 'coolwarm', annot=True,
                          annot_kws=   { 'fontsize':20, 'fontweight':'bold'   }     )

#What is {}
  # {             }
```

```python
#Plotting seaborn pairplot

plt.figure(figsize=(15,8))
sns.pairplot(data=Df, vars=['balance','income'], hue='default', palette='bright', corner=True)
plt.show()
```

```python
#Plotting seaborn Jointplot between balance and income

sns.jointplot(x='balance', y='income', data=Df, kind='hex')
plt.show()
```

```python

#Checking MAX &MIn values in
#Balance & Income columns of the DataFrame

Max_inc=Df['income'].max()
Min_inc=Df['income'].min()
print('BALANCE')
print(f'Max income: {Max_inc},  \nMin income:  {Min_inc}' )
Max_bal=Df['balance'].max()
Min_bal=Df['balance'].min()
print()
print('INCOME')
print(f'Max balance: {Max_bal},  \nMin balance:  {Min_bal}' )
```
