#GRADIENT DESCENT PROCEDURE

```python
def gradientDescent(X, y, theta, alpha, num_iters):
    """
       Performs gradient descent to learn theta
    """
    m = y.size  # number of training examples
    for i in range(num_iters):
        y_hat = np.dot(X, theta)
        theta = theta - alpha * (1.0/m) * np.dot(X.T, y_hat-y)
    return theta
```

```python
def SGD(f, theta0, alpha, num_iters):
    """
       Arguments:
       f -- the function to optimize, it takes a single argument
            and yield two outputs, a cost and the gradient
            with respect to the arguments
       theta0 -- the initial point to start SGD from
       num_iters -- total iterations to run SGD forReturn:
       theta -- the parameter value after SGD finishes
    """
    start_iter = 0
    theta= theta0
    for iter in xrange(start_iter + 1, num_iters + 1):
        _, grad = f(theta)
        theta = theta - (alpha * grad) # there is NO dot product!
    return theta
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Generate synthetic regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)
y = y.reshape(-1, 1)
m = X.shape[0]

X_b = np.c_[np.ones((m, 1)), X]

theta = np.array([[2.0], [3.0]])

plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="green", label="Initial Line (No GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()
```

```python
learning_rate = 0.1
n_iterations = 100

# Performing Gradient Descent
for _ in range(n_iterations):
    # Predictions
    y_pred = X_b.dot(theta)
    # Computing the gradient
    gradients = (2 / m) * X_b.T.dot(y_pred - y)
    # Updating parameters
    theta -= learning_rate * gradients

plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="red", label="Optimized Line (With GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression With Gradient Descent")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

df_gd= pd.DataFrame({'x1': [1,2,3,4],
                     'x2': [1,1,2,3],
                     'x3': [2,3,1,2],
                     'y': [7, 10, 12,20]})
df_gd
```

```python
# Create a Function to prepare the Data

def prepareData(data, n):
    """
    Add 1s column, convert to matrices, initialise them
    Args:
         data: read the data file
            n: int
    Return:
            x: m x (n+1) matrix
            y: m x 1 vector
        theta: (n+1) x 1 vector
    """

    # Add a column of 1s in the dataset
    data.insert(0, 'ones', 1)

    #Define X & y, seperately the dataset
    X = data.iloc[:,0:n+1]
    y = data.iloc[:,n+1:n+2]

    # convert to matrices and
    # initialize parameters theta to zero
    # theta is a vector [n+1 x 1]
    # theta Transpose is a vector [1 x n+1]
    # where n is the number of features

    X = np.matrix(X.values)
    y = np.matrix(y.values)
    theta = np.matrix(np.zeros((n+1, 1)))

    return X, y, theta
```

```python
x, y, theta = prepareData(df_gd, n)

print(x, '\n', y, '\n', theta)
```

```python
#COMPUTING THE COST

#Create a function to compute cost

def computeCost(x, y, theta):
    """
    Computing the cost function
    Args:
         x: m x n+1 matrix
         y: m x 1 vector
     theta: n+1 x 1 vector

   Returns:
      cost: float
    """
    m = len(x)
    print('m:', m)
    cost = (np.sum(np.square((x*theta)-y)))/(2*m)

    return cost
```

```python
# CALCULATING THE GRADIENT DESCENT OF THE OBJECTIVE/LOSS/COST FUNCTION

def gradientDescent(x, theta, iteration):
    """
    Implement Gradient Descent
    Args:
        x: m x n+1 matrix
    theta: n+1 x 1 vector

    Returns:
       theta: n+1 x1 vector
      J_vals: #iteration by vector
    """

    m = len(x)
    J_vals = []

    for i in range(iteration):
        error = (x * theta)-y
        for j in range(len(theta.flat)):
            theta.T[0, j] = theta.T[0, j]-((alpha/(m))*(np.sum(np.multiply(error,x[:,j]))))
            print(theta.T[0, j])

    J_vals.append(computeCost(x, y, theta))

    return (theta, J_vals)
```

```python
#overriding The Logistic Regression class methods

class LogisticRegression():

      def __init__(self, alpha, iterations):
          # learning rate, iterations
          self.alpha=alpha
          self.iterations=iterations

      def fit(self, X, Y):
          #num of samples, number of features
          self.m, self.n=Fx.shape

          #weight initialisation
          self.W= np.zeros(self.n)
          self.b= 0
          self.X=X
          self.Y=Y

          #Gradient Descent Learning
          for i in range(self.iterations):
             self.update_weights()
          return self

      def update_weights(self):
          A= 1/(1+(np.exp(-(self.X.dot(self.W)+self.b))))

          #calculate gradients
          tmp= (A-self.Y.T)
          tmp=np.reshape(tmp, self.m)
          dW= np.dot(self.X.T, tmp)/self.m
          db= np.sum(tmp)/self.m
          print("dW:", dW)
          print("db:", db)

          #update weights
          self.W=self.W-self.alpha*dW
          self.b=self.b-self.alpha*db
          return self

      def predict(self, X):
          Z= 1/(1+(np.exp(-(X.dot(self.W)+self.b))))
          Y=np.where(Z>0.5, 1,0)
          return Y
```

```python
import numpy as np
#from keras.losses import binary_crossentropy

# Example true labels and predicted probabilities
y_true = np.array([1, 0, 1, 1, 0])
y_pred = np.array([0.35, 0.23, 0.14, 0.08, 0.05])

# Compute Binary Cross-Entropy using NumPy
def binary_cross_entropy(y_true, y_pred):
    bce = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return bce

bce_loss = binary_cross_entropy(y_true, y_pred)
print(f"Binary Cross-Entropy Loss (manual calculation): {bce_loss}")

# Compute Binary Cross-Entropy using Keras
bce_loss_keras = binary_crossentropy(y_true, y_pred).numpy()
print(f"Binary Cross-Entropy Loss (Keras): {bce_loss_keras}")
```

```python

from pandas import read_csv
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
```

```python

def create_RNN(hidden_units, dense_units, input_shape, activation):
    model = Sequential()
    model.add(SimpleRNN(hidden_units, input_shape=input_shape,
                        activation=activation[0]))
    model.add(Dense(units=dense_units, activation=activation[1]))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

demo_model = create_RNN(2, 1, (3,1), activation=['linear', 'linear'])
```

```python

wx = demo_model.get_weights()[0]
wh = demo_model.get_weights()[1]
bh = demo_model.get_weights()[2]
wy = demo_model.get_weights()[3]
by = demo_model.get_weights()[4]

print('wx = ', wx)
print('wh = ', wh)
print('bh = ', bh)
print('wy = ', wy)
print('by = ', by)
```

#OLS OF STATSMODEL LIBRARY

```python
import statsmodels.api as sm
import numpy as np

# Generate some sample data
np.random.seed(0)
X = np.random.rand(100, 2)  # 2 independent variables
y = 2 * X[:, 0] + 3 * X[:, 1] + np.random.randn(100) * 0.5 # Dependent variable with noise

# Add a constant term to the independent variables (for the intercept)
X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(y, X)
results = model.fit()

# Print the results
print(results.summary())
```

#CIFAR-100 DATASET EXAMPLE

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load CIFAR-100 dataset
(cifar100_train, cifar100_test) = tf.keras.datasets.cifar100.load_data()

# Unpack the dataset
(x_train, y_train) = cifar100_train
(x_test, y_test) = cifar100_test

# Define a function to plot a grid of images
def plot_images(images, labels, class_names, grid_size=(4, 4)):
    plt.figure(figsize=(8, 8))
    for i in range(grid_size[0] * grid_size[1]):
        plt.subplot(grid_size[0], grid_size[1], i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i])
        plt.xlabel(class_names[labels[i][0]])
    plt.show()

# Get class names for CIFAR-100
class_names = [
    'apple', 'aquarium_fish', 'baby', 'bear', 'beaver', 'bed', 'bee', 'beetle', 'bicycle', 'bottle',
    'bowl', 'boy', 'bridge', 'bus', 'butterfly', 'camel', 'can', 'castle', 'caterpillar', 'cattle',
    'chair', 'chimpanzee', 'clock', 'cloud', 'cockroach', 'couch', 'crab', 'crocodile', 'cup', 'dinosaur',
    'dolphin', 'elephant', 'flatfish', 'forest', 'fox', 'girl', 'hamster', 'house', 'kangaroo', 'keyboard',
    'lamp', 'lawn_mower', 'leopard', 'lion', 'lizard', 'lobster', 'man', 'maple_tree', 'motorcycle', 'mountain',
    'mouse', 'mushroom', 'oak_tree', 'orange', 'orchid', 'otter', 'palm_tree', 'pear', 'pickup_truck', 'pine_tree',
    'plain', 'plate', 'poppy', 'porcupine', 'possum', 'rabbit', 'raccoon', 'ray', 'road', 'rocket',
    'rose', 'sea', 'seal', 'shark', 'shrew', 'skunk', 'skyscraper', 'snail', 'snake', 'spider',
    'squirrel', 'streetcar', 'sunflower', 'sweet_pepper', 'table', 'tank', 'telephone', 'television', 'tiger', 'tractor',
    'train', 'trout', 'tulip', 'turtle', 'wardrobe', 'whale', 'willow_tree', 'wolf', 'woman', 'worm'
]

# Plot a 4x4 grid of images from the training set
plot_images(x_train, y_train, class_names)
```
