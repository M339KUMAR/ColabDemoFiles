```python
dfsort=Df.sort_values(by=['total_amount'], ascending=True)
dfsort['total_amount']
```

```python

plt.figure(figsize=(20,10))
sns.lineplot(dfsort['total_amount'])
plt.xlabel('TOTAL AMOUNT', size=20)
plt.ylabel('Quantity')
plt.grid('on')
plt.legend('true')
plt.show()
```

```python
df100=Df['store'].value_counts()
labels = df100.index
print(labels)
```

```python
Counts=dfsort['quantity'].groupby(Df['store']).max()

plt.figure(figsize=(20,10))
figure=pe.bar(dfsort, x=labels, y=Counts,
              title='Plotly Bar Chart',
              #xaxis_title='STORE',
              #yaxis_title='QUANTITY',
              color= labels)

figure.show()
```
