#Plotting a Histogram with diff Parameters

```python

#https://stackoverflow.com/questions/27350226/how-to-make-the-text-size-of-the-x-and-y-axis-labels-and-the-title-on-matplotlib

import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

#https://www.geeksforgeeks.org/matplotlib-pyplot-hist-in-python/amp/
plt.hist(x, 50, density=True, facecolor='g', alpha=0.80)

plt.xlabel('Smarts', size = 20)
plt.ylabel('Probability', size=20)
plt.title('Histogram of IQ', size = 30)

plt.text(60, .025, r'$\mu=100,\ \sigma=15$', size = 20)

plt.axis([40, 160, 0, 0.03])
plt.xticks(size = 10)
plt.yticks(size = 10)

plt.grid(True)
plt.show()
```
