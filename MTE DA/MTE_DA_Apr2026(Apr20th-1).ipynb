###Task 3:

```python
import numpy as np
```

```python
def P_lh_given_A(ages_of_death, study_year= 1990):

    """
    inputs: death age & study year
    output: Prob of left handed
            given that a subject died
            in study-year at 'ages_of_death'
    """

    early_1900s_rate = lefthanded_data['Mean_lh'][-10:].mean()
    late_1900s_rate  = lefthanded_data['Mean_lh'][:10].mean()

    Y_age = 10 + (study_year - 1986)
    O_age   = 86 + (study_year - 1986)

    Mid_Rate = lefthanded_data.loc[lefthanded_data['Birth_year'].isin(study_year-ages_of_death)]['Mean_lh']

    P_return = np.zeros(ages_of_death.shape)

    P_return[ages_of_death > O_age] = early_1900s_rate/100
    P_return[ages_of_death < Y_age] = late_1900s_rate/100

    P_return[np.logical_and((ages_of_death <= O_age), (ages_of_death >= Y_age))] = Mid_Rate/100

    return P_return
```

###Task 5:

```python
def P_lh(death_distribution_data, study_year=1990):
    """
       Overall Prob of left-handedness
       in the population for a given study year

       input  : death_dist_data, study_year
       output : prob of death as lefthanded
    """

    p_list = death_distribution_data['Both Sexes']*P_lh_given_A(death_distribution_data['Age'], study_year)
    p = np.sum(p_list)

    return p/np.sum(death_distribution_data['Both Sexes'])
```

###Task 6:

```python
def P_A_given_lh(ages_of_death, death_distribution_data, study_year=1990):
    """
       P_A, The Overall Probability of
       dying at age A, which is given
       by death_distrubution_data at
       age A divided by total number of
       dead people who are left handed.

       input  : ages_of_death, study_year
       output : prob of dying at age A given lefthanded
    """

    P_A = death_distribution_data['Both Sexes'][ages_of_death]/np.sum(death_distribution_data['Both Sexes'])

    P_left = P_lh(death_distribution_data, study_year)

    P_lh_A = P_lh_given_A(ages_of_death, study_year)

    return (P_lh_A*P_A)/P_left
```

###Task 7:

```python
def P_A_given_rh(ages_of_death, death_distribution_data, study_year=1990):
    """
       The Overall Probability of
       dying at age A, which is given
       by death_distrubution_data at
       age A divided by total number of
       dead people who are right handed.

       input  : ages_of_death, study_year
       output : prob of dying at age A given lefthanded
    """

    P_A = death_distribution_data['Both Sexes'][ages_of_death]/np.sum(death_distribution_data['Both Sexes'])

    P_right = 1 - P_lh(death_distribution_data, study_year)

    P_rh_A = 1 - P_lh_given_A(ages_of_death, study_year)

    return (P_rh_A*P_A)/P_right
```

###Task 8:

```python
ages = np.arange(0, 115,1)
#print(ages)

left_handed_prob = P_A_given_lh(ages, death_distribution_data, 1990)  # for age in ages ]
right_handed_prob = P_A_given_rh(ages, death_distribution_data, 1990) #for age in ages ]

plt.plot(ages, left_handed_prob, label = "Left Handed")
plt.plot(ages, right_handed_prob, label = "Right Handed")
plt.title("Left vs Right Handed Prob", color = 'Red', size = 15)
plt.xlabel("Ages", size=15, color='green')
plt.ylabel("Probabilities at Age of Death", size= 15, color= 'green')
plt.legend()

plt.show()
```

###Task 9:

```python
Avg_lh_age = np.nansum(ages*np.array(left_handed_prob))
Avg_rh_age = np.nansum(ages*np.array(right_handed_prob))

print(round(Avg_lh_age,2))

print(round(Avg_rh_age,2))

print("The difference between the two average ages is : "+ str(round(Avg_rh_age-Avg_lh_age,2))+"Years")
```
